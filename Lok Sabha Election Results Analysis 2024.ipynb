# UP DATA

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_election_results(webpage_url):
    # Fetch the webpage content
    webpage_response = requests.get(webpage_url)

    # Parse the HTML content using beautiful soup
    soup = BeautifulSoup(webpage_response.content, 'html.parser')

    # Find the table with the required class name
    election_table = soup.find('table', class_='table table-striped table-bordered')

    #if the table is present
    if election_table:
        # Extract table headers, skipping the first header row with colspan
        table_headers = [header.get_text(strip=True) for header in election_table.find_all('th') if not header.has_attr('colspan')]

        # Initialize a list to store the rows
        table_rows = []

        # Iterate through each row in the entire table
        for table_row in election_table.find('tbody').find_all('tr'):
            row_data = []
            for table_cell in table_row.find_all('td', recursive=False):
                # Check for nested tables
                nested_table = table_cell.find('table')
                if nested_table:
                    nested_cell = nested_table.find('td')
                    row_data.append(nested_cell.get_text(strip=True) if nested_cell else "")
                else:
                    cell_content = table_cell.get_text(strip=True)
                    # Exclude unwanted rows which are not required in the data frame
                    if any(unwanted_text in cell_content for unwanted_text in ['Leading In', 'Won In', 'Trailing In']):
                        break
                    row_data.append(cell_content)
            else:
                # Only append the rows which are useful in dataframe
                table_rows.append(row_data)

        # Ensure each row has the same number of columns as the headers
        filtered_rows = [row for row in table_rows if len(row) == len(table_headers)]

        # Create a DataFrame
        election_results_df = pd.DataFrame(filtered_rows, columns=table_headers)

        return election_results_df
    else:
        print("Table not found.")
        return pd.DataFrame()  # Return an empty DataFrame


# Example usage
res = fetch_election_data("https://results.eci.gov.in/PcResultGenJune2024/statewiseS241.htm")

res_1= fetch_election_data("https://results.eci.gov.in/PcResultGenJune2024/statewiseS242.htm")

res_2= fetch_election_data("https://results.eci.gov.in/PcResultGenJune2024/statewiseS243.htm")

res_3 = fetch_election_data("https://results.eci.gov.in/PcResultGenJune2024/statewiseS244.htm")


data_up_out = pd.DataFrame(res)

data_up_1 = pd.DataFrame(res_1)

data_up_2 = pd.DataFrame(res_2)

data_up_3 = pd.DataFrame(res_3)

frames = [data_up_out , data_up_1 , data_up_2 , data_up_3]

data_up = pd.concat(frames)



data_up
```

```python
data_up.to_csv("UttarPradeshData.csv",index=False)
```
